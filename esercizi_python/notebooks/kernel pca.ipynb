```python
import numpy as np
import pandas as pd
from sklearn.datasets import make_circles, make_moons
from sklearn.decomposition import PCA, KernelPCA
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
datasetA, y_A = make_circles(n_samples=1000, factor=0.3, random_state=42)
datasetA_noise, y_A_noise = make_circles(n_samples=1000,noise=0.04,factor=0.4,random_state=42)
datasetB, y_B = make_moons(n_samples=1000, random_state=42)
datasetB_noise, y_B_noise = make_moons(n_samples=1000, noise=0.04, random_state=42)
```

```python
datasetA = pd.DataFrame(data=datasetA, columns=['x','y'])
datasetA_noise = pd.DataFrame(data=datasetA_noise, columns=['x','y'])
datasetB = pd.DataFrame(data=datasetB, columns=['x','y'])
datasetB_noise = pd.DataFrame(data=datasetB_noise, columns=['x','y'])
```

```python
sns.scatterplot(x='x', y='y',data=datasetA, hue=y_A)
```

```python
sns.scatterplot(x='x', y='y',data=datasetA_noise, hue=y_A_noise)
```

```python
sns.scatterplot(x='x', y='y',data=datasetB, hue=y_B)
```

```python
sns.scatterplot(x='x', y='y',data=datasetB_noise, hue=y_B_noise)
```

```python
KPCA = KernelPCA(n_components=2,kernel="rbf", gamma=15)
kpc_A = KPCA.fit_transform(datasetB_noise)
#invKpc_A = KPCA.inverse_transform(kpc_A)
```
